# Removing unwanted columns 

```python
data1=data.drop(columns=["ID","lat","lon"])
```

```python
data1=pd.get_dummies(data1)
```

# remove the actual value  from the dataframe

```python
y=data1['price']
x=data1.drop(columns='price')
```

# split the data into training set and testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=42)# 66% and 33%  
```

```python
x_test
```

```python
x_train
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()#creating object of LinearRegression
reg.fit(x_train,y_train)#training and fitting
```

```python
y_pred=reg.predict(x_test) #predict the price using x_test data
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score #to know the efficiency bw the predicted price
r2_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error#calaculating mse
mean_squared_error(y_test,y_pred)
```

```python
results=pd.DataFrame(columns=['Price','Predicted']) #create datafame for price and predicted
results['Price']=y_test
results['Predicted']=y_pred
```
